```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns

data = pd.read_csv("study.csv")
data = data.drop(columns="Unnamed: 0")

data["att_cat"] = data['attention'].apply(lambda x: str(x) + ' ')

sns.displot(
    data, 
    x="ror_volatility", 
    hue="attention", 
    element='step'
)
```

```python
means = data.columns[:6]
stds = data.columns[6:12]
```

```python
data[list(means) + ['attention']]
```

```python
mdata_mean = pd.melt(
    data[list(means) + ['attention']], 
    id_vars=['attention'], 
    value_vars=list(data.columns[:6]),
    var_name="stat"
)

mdata_std = pd.melt(
    data[list(stds) + ['attention']], 
    id_vars=['attention'], 
    value_vars=list(data.columns[6:12]),
    var_name="stat"
)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.lineplot(
    ax=ax,
    data=mdata_mean,
    x="attention",
    y="value",
    hue="stat")
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.lineplot(
    ax=ax,
    data=mdata_std,
    x="attention",
    y="value",
    hue="stat")
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.lineplot(
    ax=ax,
    data=data,
    x="attention",
    y="ror_volatility")
```
