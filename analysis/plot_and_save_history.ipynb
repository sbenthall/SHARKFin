```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
history = pd.read_csv('test_realistic_ammps_history.csv')
```

```python
import os
os.mkdir('figs')
```

```python
def plot_and_save(df, col):
    sns.lineplot(df.index, df[col])
    plt.savefig(f'figs/{col}.png')
```

```python
history['buy_limit'] = history['buy_sell'].apply(lambda x: eval(x)[0])
history['sell_limit'] = history['buy_sell'].apply(lambda x: eval(x)[1])
```

```python
plot_and_save(history, 'owned_shares')
```

```python
plot_and_save(history, 'total_assets')
```

```python
plot_and_save(history, 'mean_income_level')
```

```python
plot_and_save(history, 'total_consumption_level')
```

```python
plot_and_save(history, 'permshock_std')
```

```python
buy_sell = history[['Unnamed: 0', 'buy_limit', 'sell_limit']].melt('Unnamed: 0', var_name='buy_or_sell', value_name='limit')
```

```python
sns.barplot(data=buy_sell, x="Unnamed: 0", y="limit", hue="buy_or_sell")
plt.savefig('figs/buy_sell.png')
```
